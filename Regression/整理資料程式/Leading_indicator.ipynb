```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 處理領先指標資料

```python
df = pd.read_csv('/Users/Kang/Desktop/金融科技Final Project/領先指標csv/國發會領先指標.csv', index_col=0)
df = df.iloc[1:]
df = df.set_index(pd.to_datetime(df.index))
```

```python
title = df.columns.tolist()
```

```python
title
```

```python
#原本資料裡的comma無法被正確辨識，所以要先把它拿掉
for i in (3,4,6,7):
    column = title[i]
    df[column] = df[column].apply(lambda x: float(x.split()[0].replace(',', '')))
```

```python
#整理資料型態成float
df = df.astype(np.float64)
```

```python
#把資料時間點改成月底
df = df.resample('M').mean()
```

```python
df
```

```python
from matplotlib.font_manager import FontProperties
font_noto_12 = FontProperties(fname = '/Library/Fonts/NotoSansCJKtc-Regular.otf', size = 12)
font_noto_18 = FontProperties(fname = '/Library/Fonts/NotoSansCJKtc-Regular.otf', size = 18)
from IPython.core.pylabtools import figsize
figsize(12, 8)
```

```python
plt.plot(df[df.columns[0:2]], linewidth=2)
plt.title('領先指標綜合指數(點)', fontproperties = font_noto_18)
plt.legend(['含趨勢', '不含趨勢'], loc = 'best', prop = font_noto_12)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
```

```python
from matplotlib.ticker import FuncFormatter
#用來為y軸加上單位縮寫的函式
def y_fmt(y, pos):
    decades = [1e9, 1e6, 1e3, 1e0, 1e-3, 1e-6, 1e-9 ]
    suffix  = ["B", "M", "k", "" , "m" , "u", "n"  ]
    if y == 0:
        return str(0)
    for i, d in enumerate(decades):
        if np.abs(y) >=d:
            val = y/float(d)
            signf = len(str(val).split(".")[1])
            if signf == 0:
                return '{val:d} {suffix}'.format(val=int(val), suffix=suffix[i])
            else:
                if signf == 1:
                    #print (val, signf)
                    if str(val).split(".")[1] == "0":
                        return '{val:d} {suffix}'.format(val=int(round(val)), suffix=suffix[i]) 
                tx = "{"+"val:.{signf}f".format(signf = signf) +"} {suffix}"
                return tx.format(val=val, suffix=suffix[i])

                #return y
    return y
```
