```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/Kang/Desktop/individual_result.csv')
```

```python
yuanta = df[df['基金名稱'].str.contains('元大')]
cathay = df[df['基金名稱'].str.contains('國泰')]
nomura = df[df['基金名稱'].str.contains('野村')]
```
