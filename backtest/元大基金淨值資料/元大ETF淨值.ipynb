```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 整理資料

```python
df = pd.read_csv('元大基金淨值.csv', index_col = 0)
df = df.set_index(pd.to_datetime(df.index))

df1 = pd.read_csv('國泰基金淨值.csv',encoding = 'big5', index_col = 0)
df1 = df1.set_index(pd.to_datetime(df1.index))

df2 = pd.read_csv('野村基金淨值.csv', index_col = 0)
df2 = df2.set_index(pd.to_datetime(df2.index))

df3 = pd.read_csv('野村基金淨值1.csv', index_col = 0)
df3 = df3.set_index(pd.to_datetime(df3.index))

df = pd.concat([df, df1, df2, df3], axis = 1)
```

```python
#先把上下界以外資料切除
start = df.loc[(df.index.year >= 2016) & (df.index.month >= 4)].index.min()
end = df.loc[(df.index.year <= 2019) & (df.index.month <= 3)].index.max()
df = df.loc[start:end]
#把第一天沒有資料的ETF刪除
df = df[df.columns[pd.notnull(df.iloc[0])]]
```

```python
df.to_csv('complete_etf.csv', encoding = 'big5hkscs')
```
