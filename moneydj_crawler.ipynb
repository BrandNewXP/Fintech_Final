```python
from bs4 import BeautifulSoup
from io import BytesIO
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import Select
from time import sleep

import csv
import datetime
import numpy as np
import pandas as pd
import requests
import time

def waiting(b): 
    msg = b.find_element_by_id("divBlock")
    while msg.is_displayed():
        sleep(1)

#STEP 1: 呼叫Chrome的Web driver
main_driver = webdriver.Chrome('/Users/pptaira/Documents/Fintech_2019S/chromedriver') 
main_driver.implicitly_wait(20)

#STEP 2: 打開目標網頁
main_driver.get('https://www.moneydj.com/funddj/yp/yp011000.djhtm?a=ACYT148')


#STEP 3: 開始建立表格
df = pd.DataFrame(columns = ['基金名稱', '風險報酬等級', '財務報告書', '公開說明書', '基金月報', '簡式公開說明書'])
select = Select(main_driver.find_element_by_name('selTFund3'))
options = len(select.options)


for i in range(1, options + 1):
    soup = BeautifulSoup(main_driver.page_source, "html.parser")
    
    tables = soup.findAll('table')
    basic_table = pd.read_html(str(tables))[10]
    #基金名稱
    name = basic_table[1][0]
    print('Now crawling:', name)
    #風險報酬等級
    RR = basic_table[3][7]
    #財務報告書, 公開說明書, 基金月報, 簡式公開說明書的下載鏈結
    dl = [0, 0, 0, 0]
    itr = 0
    files = soup.select('a[href^="http://fundreport.funddj.com/Get"]')
    for j in (files):
        dl[itr] = j.get('href')
        itr += 1
    df.loc[i] = [name, RR, dl[0], dl[1], dl[2], dl[3]]
    if i == options + 1:
        continue
    #換下一頁
    select_rec = Select(main_driver.find_element_by_name('selTFund3'))
    select_rec.select_by_index(i)
```
