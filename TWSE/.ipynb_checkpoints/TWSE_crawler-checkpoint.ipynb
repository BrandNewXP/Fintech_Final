```python
import requests
from io import StringIO
import pandas as pd
import datetime as dt
import time

# 三大法人
dfs = []
start = dt.date(2004, 4, 7)
while start != dt.date.today():
    date = str(start).replace('-', '')
    path = 'http://www.twse.com.tw/fund/BFI82U?response=csv&dayDate=' + date + '&type=day'
    try:
        r = requests.get(path)
        df = pd.read_csv(StringIO(r.text), header=1).dropna(how='all', axis=1).dropna(how='any')
        df = df.set_index(['單位名稱'])[['買賣差額']].transpose()
        df['Date'] = start
        df.set_index(['Date'], inplace = True, drop = True) 
        dfs.append(df)
        print('{}'.format(start))
    except:
        print('{} weekend or holiday'.format(start))
    start += dt.timedelta(1)
    file = pd.concat(dfs)
#     file.to_csv('三大法人.csv', encoding = 'big5')
    time.sleep(3)
    
    
# 融資融券餘額
dfs = []
start = dt.date(2004, 4, 7)
while start != dt.date.today():
    date = str(start).replace('-', '')
    path = 'http://www.twse.com.tw/exchangeReport/MI_MARGN?response=csv&date=' + date + '&selectType=MS'
    try:
        r = requests.get(path)
        df = pd.read_csv(StringIO(r.text), skiprows = 1, error_bad_lines=False).dropna(how='all', axis=1).dropna(how='any')
        df = df.set_index(['項目'])[['今日餘額']].transpose()
        df['Date'] = start
        df.set_index(['Date'], inplace = True, drop = True)
        dfs.append(df)
        print('{}'.format(start))
    except:
        print('{} weekend or holiday'.format(start))
    start += dt.timedelta(1)
    file = pd.concat(dfs)
#     file.to_csv('融資融券餘額.csv', encoding = 'big5')
    time.sleep(3)
    

# 當日沖銷
dfs = []
start = dt.date(2014, 2, 1)
while start != dt.date.today():
# while start != dt.date(2014,4,1):
    if start.day != 1:
        start += dt.timedelta(1)
        continue
    date = str(start).replace('-', '')
    path = 'http://www.tse.com.tw/exchangeReport/TWTB4U2?response=html&date=' + date +'&stockNo='
    try:
        df = pd.read_html(path)
        dfs.append(df[0])
        print('{}'.format(start))
    except:
        print('{} weekend or holiday'.format(start))
    start += dt.timedelta(1)
    file = pd.concat(dfs)
#     file.to_csv('當日沖銷.csv', encoding = 'big5')
    time.sleep(3)
    
```
